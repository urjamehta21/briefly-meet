```python
from todoist_api_python.api import TodoistAPI
import json
import re
from datetime import datetime, timedelta

# 🔑 Replace with your API token from Todoist
api = TodoistAPI("4b76e1b37e8f2f229cd2fff57afbdfa985fb3615")

def clean_due_string(deadline_text):
    """Clean and format deadline text for Todoist API"""
    if not deadline_text:
        return None

    # Remove extra formatting and clean up
    deadline_text = deadline_text.strip().lower()
    deadline_text = re.sub(r'\\|\*', '', deadline_text)  # Remove markdown formatting

    # Handle your specific deadline formats
    if 'next meeting' in deadline_text:
        return 'next week'  # Convert to Todoist-friendly format
    elif 'before the next meeting' in deadline_text:
        return 'next week'
    elif 'today' in deadline_text:
        return 'today'
    elif 'tomorrow' in deadline_text:
        return 'tomorrow'
    elif 'next week' in deadline_text:
        return 'next week'
    elif 'next month' in deadline_text:
        return 'next month'

    # Handle date patterns (e.g., "Dec 15", "December 15, 2024", "2024-12-15")
    date_patterns = [
        r'\d{4}-\d{2}-\d{2}',  # YYYY-MM-DD
        r'\d{1,2}/\d{1,2}/\d{4}',  # MM/DD/YYYY
        r'\w+ \d{1,2}',  # Month DD
        r'\d{1,2} \w+',  # DD Month
    ]

    for pattern in date_patterns:
        if re.search(pattern, deadline_text):
            return deadline_text

    # If no recognizable date pattern, return None (no due date)
    print(f"⚠  Unrecognized deadline format: '{deadline_text}' - skipping due date")
    return None

def clean_task_content(task_text):
    """Clean task content to remove problematic characters"""
    if not task_text:
        return "Untitled Task"

    # Remove markdown formatting
    cleaned = re.sub(r'\\|\*', '', task_text)

    # Remove extra whitespace
    cleaned = ' '.join(cleaned.split())

    # Ensure it's not empty after cleaning
    if not cleaned.strip():
        return "Untitled Task"

    return cleaned.strip()

def create_todoist_tasks(json_result):
    """Create Todoist tasks from parsed meeting data"""

    if not json_result.get("action_items"):
        print("❌ No action items found in JSON result")
        return

    print(f"📝 Creating {len(json_result['action_items'])} tasks in Todoist...\n")

    successful_tasks = 0
    failed_tasks = 0

    for i, item in enumerate(json_result["action_items"], 1):
        try:
            # Clean task content first
            task_content = clean_task_content(item.get('task', f'Task {i}'))

            # Add assigned person to task content if available
            if item.get('assigned_to'):
                assigned_to = clean_task_content(item['assigned_to'])
                if assigned_to and assigned_to.lower() not in ['tbd', 'to be determined', 'unknown']:
                    task_content += f" | Assigned to: {assigned_to}"

            # Clean up deadline
            due_string = clean_due_string(item.get('deadline', ''))

            # Create task parameters
            task_params = {
                "content": task_content,
                "priority": 3  # Medium priority (1=urgent, 2=high, 3=medium, 4=low)
            }

            # Only add due_string if it's valid
            if due_string:
                task_params["due_string"] = due_string

            print(f"🔄 Creating task {i}: {task_content[:50]}...")

            # Create the task
            task = api.add_task(**task_params)

            print(f"✅ Task {i} created: {task.content}")
            if due_string:
                print(f"   📅 Due: {due_string}")
            successful_tasks += 1

        except Exception as error:
            print(f"❌ Error creating task {i}: {error}")
            print(f"   📋 Raw task: {item.get('task', 'Unknown task')}")
            print(f"   📋 Cleaned task: {clean_task_content(item.get('task', ''))}")
            print(f"   📅 Raw deadline: {item.get('deadline', 'No deadline')}")
            print(f"   📅 Cleaned deadline: {clean_due_string(item.get('deadline', ''))}")
            failed_tasks += 1

    print(f"\n📊 Summary: {successful_tasks} tasks created, {failed_tasks} failed")

    return successful_tasks, failed_tasks

# 👇 Example usage with your parsed data
# Make sure json_result is available from your previous parsing
try:
    # Verify the structure of your JSON data
    print("🔍 JSON Structure:")
    print(f"Summary items: {len(json_result.get('summary', []))}")
    print(f"Action items: {len(json_result.get('action_items', []))}")
    print()

    # Show sample action item structure
    if json_result.get('action_items'):
        print("📋 Sample action item:")
        sample_item = json_result['action_items'][0]
        for key, value in sample_item.items():
            print(f"  {key}: {value}")
        print()

    # Create tasks
    create_todoist_tasks(json_result)

except NameError:
    print("❌ json_result not found. Make sure to run the parsing code first.")
except Exception as e:
    print(f"❌ Unexpected error: {e}")

# 🛠 Alternative: Manual task creation for testing
def create_test_task():
    """Create a simple test task to verify API connection"""
    try:
        test_task = api.add_task(
            content="Test task from meeting parser",
            priority=4
        )
        print(f"✅ Test task created: {test_task.content}")
        return True
    except Exception as error:
        print(f"❌ API connection failed: {error}")
        return False

# Uncomment to test API connection
# print("🧪 Testing API connection...")
# create_test_task()
```
